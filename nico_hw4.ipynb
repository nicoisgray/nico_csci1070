```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
students_data = pd.read_csv('CSCI_Fall_2024_survey.csv')
students_data
```

```python
students_data["Student Number"] = range(len(students_data))
students_data['What state/country were you born in?'] = np.where(students_data['What state/country were you born in?'].str.contains('louis', case=False), "Missouri",
                                       np.where(students_data["What state/country were you born in?"].str.contains('mis', case=False), "Missouri",
                                                np.where(students_data['What state/country were you born in?'].str.contains('germ', case=False), "Germany",
                                                         np.where(students_data['What state/country were you born in?'].str.contains('kan', case=False), "Kansas",
                                                                  np.where(students_data['What state/country were you born in?'].str.contains('ind', case=False), "India",
                                                                          np.where(students_data['What state/country were you born in?'].str.contains('tex', case=False), "Texas", 
                                                                                   np.where(students_data['What state/country were you born in?'].str.contains('mex', case=False), "Mexico",
                                                                                            np.where(students_data['What state/country were you born in?'].str.contains('wisconsin', case=False), "Wisconsin",
                                                                                                     np.where(students_data['What state/country were you born in?'].str.contains('jersey', case=False), "New Jersey",
                                                                                                              np.where(students_data['What state/country were you born in?'].str.contains('okla', case=False), "Oklahoma",
                                                                                                                       np.where(students_data['What state/country were you born in?'].str.contains('chi', case=False), "China",
                                                                                                                                np.where(students_data['What state/country were you born in?'].str.contains('cam', case=False), "Cameroon",
                                                                                                                                         np.where(students_data['What state/country were you born in?'].str.contains('kaz', case=False), "Kazakhstan", 'Georgia')))))))))))))

#Honestly, I didn't know what half of them meant when they were listing specific breeds, or they left out the type and only put a number.
students_data['How many pets and of which kind do you have?'] = np.where(students_data['How many pets and of which kind do you have?'].str.contains('dog', case=False) & students_data['How many pets and of which kind do you have?'].str.contains('cat', case=False), "Dog and cat",
                                                                         np.where(students_data['How many pets and of which kind do you have?'].str.contains('no', case=False) | students_data['How many pets and of which kind do you have?'].str.contains('0'), "None",
                                                                                  np.where(students_data['How many pets and of which kind do you have?'].str.contains('dog', case=False), "Dog",
                                                                                           np.where(students_data['How many pets and of which kind do you have?'].str.contains('cat', case=False), "Cat", "Other/unknown"))))

students_data['What is your favorite type of food?'] = np.where(students_data['What is your favorite type of food?'].str.contains('chi', case=False) | students_data['What is your favorite type of food?'].str.contains('jap', case=False) | students_data['What is your favorite type of food?'].str.contains('kor', case=False) | students_data['What is your favorite type of food?'].str.contains('ind', case=False), "Asian",
                                                             np.where(students_data['What is your favorite type of food?'].str.contains('ita', case=False), "European",
                                                                         np.where(students_data['What is your favorite type of food?'].str.contains('mex', case=False) | students_data['What is your favorite type of food?'].str.contains('amer', case=False), "North American", "Other")))

students_data['How old are you?'] = np.where(students_data['How old are you?'] < 20, "Under 20", "Over 20")

students_data['What is your major?'] = np.where(students_data["What is your major?"].str.contains('comp', case=False), "Computer Science",
                                   np.where(students_data['What is your major?'].str.contains('business', case=False), "Business",
                                            np.where(students_data['What is your major?'].str.contains('data', case=False), "Data Science",
                                                     np.where(students_data['What is your major?'].str.contains("econ", case=False), "Economics",
                                                              np.where(students_data['What is your major?'].str.contains('cs', case=False), "Computer Science",
                                                                       np.where(students_data['What is your major?'].str.contains('bio', case=False), "Biology", "Other/Undeclared"))))))
students_data
```

```python
food_count = students_data['What is your favorite type of food?'].value_counts()
food_count.plot(kind='bar')
plt.xlabel=("Food types")
plt.ylabel=("Count")
plt.title=("Our class's favorite foods")
```

This histogram shows that most people in our class prefer Asian food.

```python
agg_age = students_data.groupby('How old are you?').count()
colors = ['red', 'blue', 'green', 'purple', 'yellow']
agg_age.plot(kind='pie', y='Student Number', colors=colors)
```

The majority of students in the class are 18 years old.

```python
plt.plot(students_data["How old are you?"], students_data["How many siblings do you have?"],'.')
```

The ages and number of siblings of our class have no relationship.

```python
birthplace = students_data["What state/country were you born in?"].value_counts()
birthplace.plot(kind="bar")
```

Most of our class is from Missouri, followed by India.

```python
agg_majors = students_data.groupby('What is your major?').count()
color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
agg_majors.plot(kind='pie', y='Student Number', colors = color)
```

Most of our class are CS majors.
