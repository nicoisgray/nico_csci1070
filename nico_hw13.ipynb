1. A neural network is a type of deep machine learning algorithm that tries to mimic the human brain and the way a human thinks. It is made of interconnected nodes (perceptrons) that process the data in layers similar to how a person makes decisions. The neural network can learn from its mistakes and improve. A neural network is made by importing a dataset to train the neural network on, choosing an activation function that allows data to move between layers, selecting the amount of layers, training the network on the dataset, and optimizing until you are satisfied with the performance.

2. A neural network's performance is evaluated with the loss function. This function measures the difference between the model's predictions and the actual values. The loss value is that difference, and the best models will have the smallest loss value possible.
