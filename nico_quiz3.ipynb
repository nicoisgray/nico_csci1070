```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df
```

```python
neighbourhoods = belize_df.groupby('neighbourhood')
neighbourhoods.first()
```

```python
nbh_price = neighbourhoods[['price']].mean()
nbh_price
```

```python
nbh_reviews = neighbourhoods[['number_of_reviews']].sum()
nbh_reviews
```

```python
nbh_available = neighbourhoods[['availability_365']].max()
nbh_available
```

```python
nbh_partial = pd.merge(nbh_price, nbh_reviews, on=['neighbourhood'])
nbh_partial
```

```python
nbh_final = pd.merge(nbh_partial, nbh_available, on=['neighbourhood'])
nbh_final
```
